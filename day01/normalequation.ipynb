```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mylinearregression import MyLinearRegression as MyLR
```

```python
data = pd.read_csv("./resources/spacecraft_data.csv")
```

```python
X = np.array(data[['Age','Thrust_power','Terameters']])
X_age = np.array(data['Age']).reshape((-1,1))
print(X.shape)
Y = np.array(data['Sell_price']).reshape((-1,1))
```

```python
myLR_ne = MyLR([1., 1., 1., 1.])
myLR_lgd = MyLR([1., 1., 1., 1.])
```

```python
myLR_lgd.fit_(X,Y, alpha = 5e-5, n_cycle = 10000)
```

```python
myLR_ne.normalequation_(X,Y)
```

```python
myLR_lgd.mse_(X,Y)
```

```python
myLR_ne.mse_(X,Y)
```

```python
print(myLR_lgd.theta.shape)
```

```python
plt.xlabel('x1 = age(in years)')
plt.ylabel('Y: selling price(in keuroons)')
plt.plot(X_age,Y,'o--')
plt.plot(X_age,myLR_lgd.predict_(X),marker='o',linestyle='None')
plt.plot(X_age,myLR_ne.predict_(X),marker='o',linestyle='None')

plt.show()
```
